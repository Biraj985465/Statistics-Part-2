3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pmf(x, p):
    """
    Calculate the PMF of a Bernoulli distribution.

    Parameters:
        x (int): 0 or 1 (outcome)
        p (float): probability of success (0 <= p <= 1)

    Returns:
        float: probability of outcome x
    """
    if x not in (0, 1):
        raise ValueError("x must be 0 or 1 for Bernoulli distribution")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1")

    return p**x * (1 - p)**(1 - x)

p_success = 0.7
print(f"P(X=1) = {bernoulli_pmf(1, p_success):.3f}")
print(f"P(X=0) = {bernoulli_pmf(0, p_success):.3f}")
```

9. Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores of the input data and plot them.

    Parameters:
        data (list or np.array): Numeric dataset

    Returns:
        z_scores (np.array): Array of Z-scores
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    z_scores = (data - mean) / std

    plt.scatter(range(len(data)), z_scores, color='blue')
    plt.axhline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.axhline(1, color='green', linestyle='--', label='Z=1')
    plt.axhline(-1, color='green', linestyle='--', label='Z=-1')
    plt.axhline(2, color='orange', linestyle='--', label='Z=2')
    plt.axhline(-2, color='orange', linestyle='--', label='Z=-2')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.title('Z-scores of the Dataset')
    plt.legend()
    plt.grid(True)
    plt.show()

    return z_scores

sample_data = [10, 12, 15, 14, 13, 9, 11, 16, 18, 14]
z_scores = calculate_and_plot_z_scores(sample_data)
print("Z-scores:", z_scores)
```

14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data, x):
    """
    Calculate the Z-score of a data point x in the dataset.

    Parameters:
        data (list or np.array): Dataset
        x (float): Data point to calculate Z-score for

    Returns:
        float: Z-score of x
    """
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    z = (x - mean) / std
    return z

def plot_z_on_standard_normal(z):
    """
    Plot the standard normal distribution and mark the Z-score.

    Parameters:
        z (float): Z-score to mark on the plot
    """
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.plot(x, y, label='Standard Normal Distribution')
    plt.fill_between(x, y, alpha=0.2)

    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
    plt.scatter(z, norm.pdf(z), color='red')

    plt.title('Z-score on Standard Normal Distribution')
    plt.xlabel('Z')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

data = [12, 15, 14, 10, 18, 20, 22, 19, 16, 14]
data_point = 18

z_score = calculate_z_score(data, data_point)
print(f"Data point: {data_point}")
print(f"Z-score: {z_score:.2f}")

plot_z_on_standard_normal(z_score)
```

15.  Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Perform a one-sample Z-test.

    Parameters:
        sample (list or np.array): sample data
        pop_mean (float): population mean under the null hypothesis
        pop_std (float): known population standard deviation
        alpha (float): significance level (default 0.05)

    Returns:
        z_stat (float): calculated Z statistic
        p_value (float): two-tailed p-value
        reject_null (bool): whether to reject the null hypothesis
    """
    sample = np.array(sample)
    n = len(sample)
    sample_mean = np.mean(sample)

    z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    reject_null = p_value < alpha

    return z_stat, p_value, reject_null

sample_data = [102, 100, 98, 97, 105, 101, 99, 103, 104, 96]
population_mean = 100
population_std = 5
alpha = 0.05

z_stat, p_value, reject = z_test(sample_data, population_mean, population_std, alpha)

print(f"Z statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.4f}")
print(f"Reject null hypothesis? {'Yes' if reject else 'No'}")
```

16. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of the data.

    Parameters:
        data (list or np.array): Sample data
        confidence (float): Confidence level (default 0.95)

    Returns:
        (float, float): Lower and upper bounds of the confidence interval
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)

    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin_error = t_crit * std_err
    lower = mean - margin_error
    upper = mean + margin_error

    return lower, upper

sample_data = [12, 15, 14, 10, 18, 20, 22, 19, 16, 14]

lower_bound, upper_bound = confidence_interval(sample_data, confidence=0.95)
mean_value = np.mean(sample_data)

print(f"Sample Mean: {mean_value:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
print(f"Interpretation: We are 95% confident that the true population mean lies within this interval.")
```

17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

```python
import numpy as np
from scipy import stats

np.random.seed(42)
mu = 100
sigma = 15
sample_size = 50

data = np.random.normal(mu, sigma, sample_size)

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)
    margin_error = t_crit * std_err
    lower = mean - margin_error
    upper = mean + margin_error
    return mean, lower, upper

mean, lower_bound, upper_bound = confidence_interval(data)

print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
print(f"Interpretation: We are 95% confident that the true population mean lies between {lower_bound:.2f} and {upper_bound:.2f}.")
```

24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(mu=0, sigma=1, size=1000):
    """
    Generate a normal distribution dataset and calculate skewness and kurtosis.

    Parameters:
        mu (float): Mean of the normal distribution.
        sigma (float): Standard deviation of the normal distribution.
        size (int): Number of data points.

    Returns:
        tuple: (skewness, kurtosis)
    """

    data = np.random.normal(loc=mu, scale=sigma, size=size)

    skewness = skew(data)

    kurt = kurtosis(data)

    return skewness, kurt

skewness, kurt = calculate_skewness_kurtosis(mu=0, sigma=1, size=1000)
print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis: {kurt:.4f}  (Note: For normal distribution, kurtosis ~ 0)")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(42)
sample_size = 50
sample = np.random.normal(loc=103, scale=15, size=sample_size)

population_mean = 100
population_std = 15
alpha = 0.05

sample_mean = np.mean(sample)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

z_critical = norm.ppf(1 - alpha/2)

print("Sample Mean:", sample_mean)
print("Z-score:", z_score)
print("P-value:", p_value)

if abs(z_score) > z_critical:
    print("Reject null hypothesis: Sample mean is significantly different.")
else:
    print("Fail to reject null hypothesis: No significant difference.")

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution')
plt.fill_between(x, y, where=(x < -z_critical) | (x > z_critical), color='red', alpha=0.3, label='Rejection Region (Œ±=0.05)')
plt.axvline(z_score, color='blue', linestyle='--', label=f'Z-score = {z_score:.2f}')
plt.axvline(-z_critical, color='black', linestyle=':', label=f'Z-critical = ¬±{z_critical:.2f}')
plt.axvline(z_critical, color='black', linestyle=':')

plt.title('Two-Tailed Z-Test: Decision Regions')
plt.xlabel('Z-value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

  5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0, mu1, sigma, n, alpha=0.05, test_type='right'):
    """
    Visualize Type I and Type II errors for a one-tailed hypothesis test.

    Parameters:
        mu0: float - Null hypothesis mean
        mu1: float - Alternative hypothesis mean
        sigma: float - Population standard deviation
        n: int - Sample size
        alpha: float - Significance level (Type I error)
        test_type: str - 'right' or 'left' for one-tailed test
    """
    se = sigma / np.sqrt(n)

    if test_type == 'right':
        z_crit = norm.ppf(1 - alpha)
        crit_val = mu0 + z_crit * se
        beta = norm.cdf((crit_val - mu1) / se)
    else:
        z_crit = norm.ppf(alpha)
        crit_val = mu0 + z_crit * se
        beta = 1 - norm.cdf((crit_val - mu1) / se)

    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    y_null = norm.pdf(x, mu0, se)
    y_alt = norm.pdf(x, mu1, se)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y_null, label='Null Hypothesis (H‚ÇÄ)', color='blue')
    plt.plot(x, y_alt, label='Alternative Hypothesis (H‚ÇÅ)', color='green')

    if test_type == 'right':
        plt.fill_between(x, 0, y_null, where=(x > crit_val), color='red', alpha=0.3, label='Type I Error (Œ±)')
        plt.fill_between(x, 0, y_alt, where=(x < crit_val), color='orange', alpha=0.3, label='Type II Error (Œ≤)')
    else:
        plt.fill_between(x, 0, y_null, where=(x < crit_val), color='red', alpha=0.3, label='Type I Error (Œ±)')
        plt.fill_between(x, 0, y_alt, where=(x > crit_val), color='orange', alpha=0.3, label='Type II Error (Œ≤)')

    plt.axvline(crit_val, color='black', linestyle='--', label=f'Critical Value = {crit_val:.2f}')

    plt.title("Type I and Type II Errors Visualization")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Critical value: {crit_val:.2f}")
    print(f"Type I error (Œ±): {alpha}")
    print(f"Type II error (Œ≤): {beta:.3f}")
    print(f"Power of the test (1 - Œ≤): {1 - beta:.3f}")
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of a sample.

    Parameters:
        data (list or array): The sample data
        confidence (float): Confidence level (default is 0.95 for 95%)

    Returns:
        (mean, lower_bound, upper_bound)
    """
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * sem
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    return mean, lower, upper

sample_data = [72, 75, 78, 74, 77, 73, 76, 79, 75, 74]
mean, lower, upper = confidence_interval(sample_data, confidence=0.95)

print(f"Sample Mean = {mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

10.  Write a Python program to calculate the margin of error for a given confidence level using sample data.

```python
import numpy as np
from scipy import stats

def margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for the sample mean.

    Parameters:
        data (list or array): Sample data
        confidence (float): Confidence level (default 0.95)

    Returns:
        margin_of_error (float)
    """
    n = len(data)
    sem = stats.sem(data)
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)
    moe = t_crit * sem
    return moe

sample_data = [85, 88, 90, 86, 87, 89, 91, 90, 88, 86]
moe = margin_of_error(sample_data, confidence=0.95)

print(f"Margin of Error (95% confidence): {moe:.3f}")
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayes_theorem(prior, likelihood, false_positive_rate):
    """
    Compute posterior probability using Bayes' theorem.

    Parameters:
        prior: P(H) - prior probability of hypothesis (disease)
        likelihood: P(E|H) - probability of evidence given hypothesis (true positive rate)
        false_positive_rate: P(E|¬¨H) - probability of evidence given not hypothesis (false positive rate)

    Returns:
        posterior: P(H|E) - updated probability after observing evidence
    """

    marginal_likelihood = (likelihood * prior) + (false_positive_rate * (1 - prior))

    posterior = (likelihood * prior) / marginal_likelihood
    return posterior

prior_disease = 0.001
true_positive_rate = 0.99
false_positive_rate = 0.05

posterior_prob = bayes_theorem(prior_disease, true_positive_rate, false_positive_rate)

print(f"Posterior probability of having disease given positive test: {posterior_prob:.4f}")
```

13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

```python
import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test from observed data.

    Parameters:
        observed (2D array): observed frequencies (contingency table)

    Returns:
        expected (2D array): expected frequencies under independence
    """
    observed = np.array(observed)
    row_totals = observed.sum(axis=1).reshape(-1, 1)
    col_totals = observed.sum(axis=0).reshape(1, -1)
    grand_total = observed.sum()

    expected = row_totals @ col_totals / grand_total
    return expected


observed_data = [
    [30, 10],  # Row 1
    [20, 40]   # Row 2
]

expected_freq = calculate_expected_frequencies(observed_data)

print("Observed frequencies:")
print(np.array(observed_data))
print("\nExpected frequencies under independence:")
print(np.round(expected_freq, 2))
```
